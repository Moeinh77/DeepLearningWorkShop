### Loading the data
keras can load the data itself without a need for extra procedures

```python
from keras.datasets import boston_housing
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
```

```python
print(x_train.shape)
print(x_test.shape)

x_train[0]
```

```python
y_train[0]
```

### Scaling the data for faster convergence

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

X_train=scaler.fit_transform(x_train)
X_test=scaler.transform(x_test)
```

### Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.utils import np_utils
```

### Designing the model

```python
model =Sequential()
model.add(Dense(64, activation='relu', input_shape=(13,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(loss='mean_squared_error', optimizer='rmsprop',metrics=['mae'])
history=model.fit(X_train, y_train,batch_size=64, epochs=1000,verbose=1,validation_split=0.2)
```

### Evaluating the model

Mean absolute error

```python
print(model.evaluate(X_test,y_test)[1])
```

```python
print(model.predict(X_test[0,:].reshape(1,13)))
print(y_test[0])
```

### How much predictions matches the real value

```python
plt.plot(y_test)
plt.plot(model.predict(X_test))
```

As you can see the model mostly makes predictions near the real value ,we can improve the predictions by trainig the model longer

```python
model.save('housing_weights.h5')
```
